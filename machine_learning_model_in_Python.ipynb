# My first ML code


# **Load data**


```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df

```

#**Data Preparation**

###Data seperation as X and Y

```python
y=df['logS']
y
```

```python
x= df.drop('logS', axis=1)
x
```

###Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

#**Model Building**

***

###**Linear regression**

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(x_train, y_train)
```

####**Training the model**

####**Apply the model to make the prediction**

```python
y_lr_train_pred= lr.predict(x_train)
y_lr_test_pred= lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

####**Evaluate model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse= mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2= r2_score(y_train, y_lr_train_pred)

lr_test_mse= mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2= r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame([['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]],
                          columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])
```

```python
# Display the DataFrame
lr_results
```

##**Random Forest**


###**Training the model**

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

###**Applying the model to make the prediction**

```python
y_rf_train_pred= rf.predict(x_train)
y_rf_test_pred= rf.predict(x_test)
```

###**Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse= mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2= r2_score(y_train, y_rf_train_pred)

rf_test_mse= mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2= r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame([['Random Forest', rf_train_mse, lr_train_r2, rf_test_mse, rf_test_r2]],
                          columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])
```

```python
rf_results
```

##**Model Comparision**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)

df_models
```

```python
df_models.reset_index()
```

##**Data Visualization of predicred results**

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting actual vs predicted for training set
plt.figure(figsize=(5, 5))
plt.scatter(y_train, y_lr_train_pred, color='#FCAB00', alpha=0.3)

# Fit and plot regression line
x = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(x)
plt.plot(y_train, p(y_train), color='#970C66')

# Labels and title
plt.xlabel('Experimental Logs')
plt.ylabel('Predicted Logs')
plt.title('Actual vs Predicted (Training Set)')
plt.grid(True)
plt.show()
```
